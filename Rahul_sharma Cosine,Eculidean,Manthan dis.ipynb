```python
##Que - 1).Cosine Similarity
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import spacy

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Define your two sentences
sentence1 =  "I am pursuing MSC from school of data science"
sentence2 = "right now school of data science is running these course"

# Tokenize and lemmatize the sentences using spaCy
tokens1 = [token.lemma_ for token in nlp(sentence1) if not token.is_punct and not token.is_space]
tokens2 = [token.lemma_ for token in nlp(sentence2) if not token.is_punct and not token.is_space]

# Combine the tokens back into sentences
processed_sentence1 = ' '.join(tokens1)
processed_sentence2 = ' '.join(tokens2)

# Create a CountVectorizer to convert the sentences to vectors
vectorizer = CountVectorizer().fit_transform([processed_sentence1, processed_sentence2])

# Calculate the cosine similarity
cosine_sim = cosine_similarity(vectorizer[0], vectorizer[1])

print(f"Cosine Similarity: {cosine_sim[0][0]}")
```

```python
##Que - 2). cosine_distance
cosine_distance = 1 - cosine_sim
##print(cosine_distance)
print(f"Cosine Distance: {cosine_distance}")

```

```python
##Que - 3).Euclidean Distance
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import pairwise_distances
import numpy as np

# Define your two sentences
sentence1 ="I am pursuing MSC from school of data science"
sentence2 = "right now school of data science is running these course"

# Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the sentences into TF-IDF vectors
tfidf_vectors = vectorizer.fit_transform([sentence1, sentence2])

# Convert the sparse TF-IDF vectors to dense arrays
dense_vectors = tfidf_vectors.toarray()

# Calculate the Euclidean distance
euclidean_distance = np.linalg.norm(dense_vectors[0] - dense_vectors[1])

print(f"Euclidean Distance: {euclidean_distance}")
```

```python
## Que - 4). manhattan distance
doc_1 ="I am pursuing MSC from school of data science"
doc_2= "right now school of data science is running these course"

# Vector representation of the document
doc_1_vector = [1, 1, 1, 1, 0, 1, 1, 2]
doc_2_vector = [1, 0, 0, 1, 1, 0, 1, 0]
data = [doc_1, doc_2]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer()
vector_matrix = count_vectorizer.fit_transform(data)
vector_matrix
```

```python
tokens = count_vectorizer.get_feature_names_out()
tokens
```

```python
vector_matrix.toarray()
```

```python
import pandas as pd

def create_dataframe(matrix, tokens):

    doc_names = [f'doc_{i+1}' for i, _ in enumerate(matrix)]
    df = pd.DataFrame(data=matrix, index=doc_names, columns=tokens)
    return(df)
```

```python
create_dataframe(vector_matrix.toarray(),tokens)
```

```python

from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity_matrix = cosine_similarity(vector_matrix)
create_dataframe(cosine_similarity_matrix,['doc_1','doc_2'])
```

```python
# computing the manhattan distance
from scipy.spatial import distance
manhattan_distance = distance.cityblock(doc_1_vector, doc_2_vector)
print('Manhattan Distance b/w', doc_1_vector, 'and', doc_2_vector, 'is: ', manhattan_distance)
```
